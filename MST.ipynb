The following code imports the necessary libraries and sets up a weighted graph using NetworkX. 

Here's a step-by-step explanation:

1. `import networkx as nx` - This line imports the NetworkX library, which is used for the creation, manipulation, and study of complex networks (graphs).

2. `import matplotlib.pyplot as plt` - This line imports the Matplotlib library's `pyplot` module, which is used for creating static, animated, and interactive visualizations in Python.

3. `G = nx.Graph()` - This line initializes an empty undirected graph `G` using NetworkX.

4. `edges = [('A', 'B', 1), ('A', 'C', 3), ('B', 'C', 1), ('B', 'D', 6), ('C', 'D', 2)]` - This line defines a list of edges for the graph, where each edge is represented as a tuple containing two nodes and a weight (the third element).

5. `G.add_weighted_edges_from(edges)` - This line adds the edges defined in the `edges` list to the graph `G`, incorporating the weights associated with each edge.

Overall, this selected code sets up the necessary components to create and visualize a weighted graph using NetworkX and Matplotlib.

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
edges = [('A', 'B', 1), ('A', 'C', 3), ('B', 'C', 1), ('B', 'D', 6), ('C', 'D', 2)]
G.add_weighted_edges_from(edges)

# Draw the weighted graph
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw(G, pos, with_labels=True)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)


plt.title("Weighted Graph from Edge List")
plt.show()
```

The following code imports necessary libraries and sets up a weighted graph using a weighted adjacency matrix. Here’s a step-by-step explanation:

1. `import networkx as nx` - This line imports the NetworkX library, which is used for the creation, manipulation, and study of complex networks.

2. `import numpy as np` - This imports the NumPy library, which is used for numerical operations and handling arrays.

3. `import matplotlib.pyplot as plt` - This imports the Matplotlib library, specifically the `pyplot` module, which is used for creating static, animated, and interactive visualizations in Python.

4. `weighted_adjacency_matrix = np.array([[0, 2, 0, 0], ...])` - This line defines a weighted adjacency matrix using a NumPy array. The matrix represents a graph where the value at position (i, j) indicates the weight of the edge between node i and node j. A value of 0 indicates no edge exists.

5. `G = nx.from_numpy_array(weighted_adjacency_matrix)` - This line creates a weighted graph `G` from the defined adjacency matrix using the `from_numpy_array` function from NetworkX. The graph will have nodes corresponding to the rows/columns of the matrix and edges with weights as specified in the matrix.

Overall, this selected code sets up the necessary libraries and data structures to create and visualize a weighted graph based on the provided adjacency matrix.

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

# Define the weighted adjacency matrix
# 0 indicates no edge, positive values indicate weights
weighted_adjacency_matrix = np.array([[0, 2, 0, 0],
                                       [2, 0, 3, 6],
                                       [0, 3, 0, 4],
                                       [0, 6, 4, 0]])

# Create a weighted graph from the adjacency matrix
G = nx.from_numpy_array(weighted_adjacency_matrix)

# Draw the weighted graph
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='blue', node_size=300)

# Draw edge labels to show weights
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Weighted Graph from Adjacency Matrix")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()
edges = [(1, 2, 1), (1, 3, 3), (2, 3, 1), (2, 4, 6), (3, 4, 2)]
G.add_weighted_edges_from(edges)

# Compute the minimum spanning tree using Kruskal's algorithm
mst = nx.minimum_spanning_tree(G)

# Draw the original graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True,  edge_color='gray', node_size=700)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

# Draw the minimum spanning tree
nx.draw(mst, pos, with_labels=True,  edge_color='red', node_size=700)
mst_labels = nx.get_edge_attributes(mst, 'weight')
nx.draw_networkx_edge_labels(mst, pos, edge_labels=mst_labels)

# Show the plot
plt.title("Minimum Spanning Tree")
plt.show()
```
